Importing the libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading the dataset

```python
url = 'https://raw.githubusercontent.com/VimalChamyal/pb_data/main/Data.csv'
```

```python
df = pd.read_csv(url)
df.head()
```

Data Preprocessing

```python
df['Profile'] = df['Profile'].str.replace('ProfileID ', '')
df['Secondary Researcher'] = df['Secondary Researcher'].str.replace('Researcher ', '')
df.head(10)
```

Summarizing the data

```python
df.info()
```

```python
df.describe()
```

Checking for duplicates

```python
df.duplicated().sum()
```

Null values

```python
df.isnull().sum()
```

Unique values in each column

```python
n = df.nunique(axis=0)

print("No.of.unique values in each column :\n", n)
```

```python
df['Workflow Status'].value_counts().plot(kind="bar")
plt.title("Workflow Status Dist")
plt.xlabel("Workflow Status")
plt.ylabel("Count")

plt.show()
```

```python
df['Workflow Priority'].value_counts().plot(kind="bar")
plt.title("Workflow Priority Dist")
plt.xlabel("Workflow Priority")
plt.ylabel("Count")

plt.show()
```

```python
df['Workflow Region'].value_counts().plot(kind="bar")
plt.title("Workflow Region Dist")
plt.xlabel("Workflow Region")
plt.ylabel("Count")

plt.show()
```

```python
sns.displot(df['Benchmark Points'])
plt.xlim(0,120)
plt.show()
```

```python
sns.displot(df['Time taken to Complete (secs)'])
plt.xlim(0,12500)
plt.show()
```

```python
sns.scatterplot(data = df, x = 'Benchmark Points', y = 'Time taken to Complete (secs)')
plt.title('Benchmark Points vs Time Taken')
plt.show()
```
